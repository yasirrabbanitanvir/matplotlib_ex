```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [50,60,52,55,62]

plt.plot(x,y, color = "red", linewidth = 5, linestyle = 'dotted')

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather Chart")
```

```python
days = [1,2,3,4,5,6]
max_t = [ 50, 51, 52, 53, 54, 48]
min_t = [42,41,43, 40, 38, 33]
avg_t= [42,41,43,40, 35, 33]

plt.plot(days, max_t, label="Max")
plt.plot(days, min_t, label="Min")
plt.plot(days, avg_t, label="Avg")
plt.legend(loc="best", shadow = True)
plt.grid()
plt.savefig("plotchart_grid.png")
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
profit = [5,7,10,5]

plt.bar(product,price, label = "price", color = 'red')
plt.bar(product, profit, label = "profit")
plt.savefig("piechart_below_shape.png")
```

```python
import numpy as np
```

```python
product = np.array([1, 2, 3, 4])
price = [15,25,40,30]
profit = [5,7,10,5]

plt.bar(product - 0.2,price, label = "price", color = 'red', width = .4)
plt.bar(product + 0.3, profit, label = "profit", width = .4)
plt.legend()
plt.savefig("barchart_different_shape.png")
```

```python
val = [300,500,1600,450, 360]
lab = ['price', 'sell', 'total', 'shirt', 'saree']
plt.pie(val, labels= lab, radius = 1, autopct = "%.2f%%", shadow = True, startangle = 45)
plt.savefig("piechart_normal_shape.png")
```

```python
val = [300,500,1600,450,360]
lab = ['price', 'sell', 'total', 'shirt', 'saree']
plt.pie(val, labels= lab, radius = 1, autopct = "%.2f%%", shadow = True, startangle = 45,explode=[0,.2,0,0,0])
plt.savefig("piechart_not_normal_shape.png")
```
